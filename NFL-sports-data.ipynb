```python
import json
import requests
from pprint import pprint
```

```python
## team schedule for 2020

url = "https://fly.sportsdata.io/v3/nfl/scores/json/Schedules/2020?key=f47e6e2ba5494c6189d60a93cc7ea7d1"

    
print(url)
```

```python
response = requests.get(url).json()
```

```python
pprint(response)
```

```python
with open("./static/data/team-schedule.json", "w") as json_file:
    json.dump(response, json_file)
```
